```python
import cv2
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# -------- PREPROCESSING --------
def preprocess(image_path):
    img = cv2.imread(image_path, 0)  # grayscale
    img = cv2.resize(img, (100, 100))
    img = cv2.GaussianBlur(img, (5,5), 0)
    return img

# -------- FEATURE EXTRACTION --------
def extract_features(img):
    return img.flatten()

# -------- LOAD DATASET --------
X = []
y = []

# Genuine signatures (label = 1)
for file in os.listdir("dataset/genuine"):
    path = os.path.join("dataset/genuine", file)
    img = preprocess(path)
    X.append(extract_features(img))
    y.append(1)

# Forged signatures (label = 0)
for file in os.listdir("dataset/forged"):
    path = os.path.join("dataset/forged", file)
    img = preprocess(path)
    X.append(extract_features(img))
    y.append(0)

X = np.array(X)
y = np.array(y)

# -------- TRAIN TEST SPLIT --------
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# -------- MODEL --------
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)

# -------- EVALUATION --------
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# -------- PREDICTION --------
test_img = preprocess("test_signature.png")
test_feature = extract_features(test_img)

prediction = model.predict([test_feature])

if prediction == 1:
    print("✅ Genuine Signature")
else:
    print("❌ Forged Signature")
```
